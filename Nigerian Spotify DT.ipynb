### What are the top 10 genres?

```python
# Group DF Data by their Genre, Count All their values each, and Sort them through Descending Order
top_genres = df.groupby("artist_top_genre").count().sort_values("name", ascending=False).head(10)
top_genres
```

From the Plot Below we see that Afro DanceHall has the highest data. Missing data accounts for artists who input their own "form of genre", that may haven't been recognised by Spotify

```python
plt.figure(figsize=[8,8])
plt.title("Top 10 Genres")

sns.barplot(y="artist_top_genre", x="name", data=top_genres.reset_index())
plt.xlabel("Count");
```

### Top 20 Artists with the Highest number of songs?

```python
artist_no_of_songs = df.groupby("artist").count().sort_values("name",ascending=False).head(20)
artist_no_of_songs
```

We see that Various Artists stands out, this are Artistes who did collaborations with each other to make the songs

```python
plt.figure(figsize=[8,8])
plt.title("Artistes with the Highest No of Songs")

# Resetting Index so that the Artist Index would be a column itself
sns.barplot(y="artist", x="name", data=artist_no_of_songs.reset_index())
plt.xlabel("Count");
```

### What are the top 20 albums with the highest songs?

```python
# Top Albums
album_no_of_songs = df.groupby("album").count().sort_values("name",ascending=False).head(20)
```

```python
plt.figure(figsize=[8,8])
plt.title("Albums with the Highest No of Songs")

sns.barplot(y="album", x="name", data=album_no_of_songs.reset_index())
plt.xlabel("Count");
```

### What are the top 30 Most Danceable Songs?

```python
# Most Danceable songs
danceable_songs = df.sort_values("danceability", ascending=False).head(30)
danceable_songs
```

```python
plt.figure(figsize=[8,14])
plt.title("Top 30 Most Danceable Songs")

sns.barplot(x="danceability", y="name", data=danceable_songs);
```

### Do Songs Need to Have HIgh Energy to be Danceable?

From the Plot below is Almost No, as some Danceable Songs have low energy, but some of the most danceable songs are within average energy

```python
# Songs with High Danceability dont all need to have high energy
plt.figure(figsize=[8,6])
plt.title("Energy Against Danceabilty")
sns.scatterplot(x="danceability", y="energy", data=df.head(30));
```

### Do Songs Need to Have Long to be Popular?

From the Plot below is the answer is No, As Some Popular Songs are below 220000 milliseconds

```python
# Songs dont need to be really long to be popular
plt.figure(figsize=[8,6])
plt.title("Length Vs Popularity")

top_30_songs = df.sort_values("popularity",ascending=False).head(30)
sns.scatterplot(x="popularity", y="length", data=top_30_songs)

plt.ylabel("Length (milliseconds)");
```
